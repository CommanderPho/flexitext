```python
import matplotlib.pyplot as plt
from flexitext import FlexiText, Style
```

```python
s1 = Style(size=28, color="blueviolet", name="Montserrat")
s2 = Style(color="grey", size=24)
s3 = Style(color="royalblue", size=24, style="italic", name="Montserrat")
s4 = Style(color="royalblue", size=24, name="Montserrat")
s5 = Style(color="royalblue", size=24, weight="bold", name="Montserrat")

ft = FlexiText(
    s1("You can write using\n"), 
    s2("multiple formats,\nand linebreaks\n\n"), 
    s4("also "), s5("bold text"), s4("\n\n"),
    s2("and why not "), s3("italics"), s2(" too")
)

fig, ax = plt.subplots(figsize=(9, 6))
ax.axvline(0.3)
ft.plot(0.3, 0.6, ha="left", ma="left");
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.axhline(0.6)
ft.plot(0.3, 0.6, va="top", ha="left", ma="left");
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.axhline(0.3)
ft.plot(0.3, 0.3, va="bottom", ha="left", ma="left");
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ft.plot(0.5, 0.5, va="center", ha="center", ma="center");
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
fig.subplots_adjust(top=0.8, left=0.025)

ft = FlexiText(
    s1("Also outside of the Axis\n"), 
    s2("Using figure fraction coordinates!"),
)
ft.plot(0.025, 0.8, va="bottom", xycoords="figure fraction")

fig.set_facecolor("white")
fig.savefig("plot.png")
```
